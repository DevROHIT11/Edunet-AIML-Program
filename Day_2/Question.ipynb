# Question -- Find city greater than 26 deg temp and avg of carbon footprint

```python
climate_data = [
    {"city": "City A", "temperature": 25, "carbon_footprint": 500},
    {"city": "City B", "temperature": 30, "carbon_footprint": 350},
    {"city": "City C", "temperature": 22, "carbon_footprint": 600},
    {"city": "City D", "temperature": 15, "carbon_footprint": 200},
    {"city": "City E", "temperature": 28, "carbon_footprint": 450},
]

# High temperature
sum = 0
print("Cities with high temperature than 26")
for data in climate_data:
    if data["temperature"] > 26 :
        print(data["city"])
        
    sum += data["carbon_footprint"]
    avg = sum / len(climate_data)
  
print(f"The average is {avg}")

# sustainable cities
def Sustainable(climate_data):
    for data in climate_data:
        if(data["carbon_footprint"] < avg):
            print(f"The Sustainanble city is {data["city"]}")

# Non- sustainable cities
def NonSustainable(climate_data):
     for data in climate_data:
        if(data["carbon_footprint"] > avg):
            print(f"The Sustainanble city is {data["city"]}")

# calling the function
Sustainable(climate_data)
NonSustainable(climate_data)

#highest carbon footprint value 
# def Highest(data):
#     for data in climate_data:
#         if(data["carbon_footprint"] > ):
#             print(f"The Sustainanble city is {data["city"]}")
```
