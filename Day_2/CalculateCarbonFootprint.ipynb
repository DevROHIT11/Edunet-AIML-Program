```python
def CarbonFootprint(energy , total_waste, ef_electricity , ef_waste):
    ec_cf = energy * ef_electricity
    tw_cf = total_waste + ef_waste

    CarbonFootprint = ec_cf + tw_cf

    print(f"The Carbon Footprint is = {CarbonFootprint} KW-Hrs")
    # return ec_cf + tw_cf


energy = 1000
total_waste = 6800

CarbonFootprint(energy , total_waste , 0.475 , 0.92)
```
