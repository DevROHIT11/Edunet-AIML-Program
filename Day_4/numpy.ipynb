### NUMPY

```python
import numpy as np 

# converting an list to array
ec = [1200 , 3400 , 2900 , 1800, 2500]
energy_consumption = np.array(ec)

print("Energy consumption (in MWh) for different Renewable sources:")
print(energy_consumption)
```

#### 1. CREATING MATRIX

```python
# 1d array
A = np.ones(4)
A2 = np.ones((2,2))

# 2d array
print(A)
print(A2)

# array with zeros
M = np.zeros(4)
print(M)
```

#### 2. IDENTITY MATRIX

```python
i = np.identity(4)
print(i)
```

#### 3. Sum of array elements

```python
ec = [1200 , 3400 , 2900 , 1800, 2500]
energy_consumption = np.array(ec)

total = np.sum(energy_consumption)
print("The total oonsumption is :" ,total)
```

#### 4. MEAN

```python
mean = np.mean(energy_consumption)
print("The mean is :", mean)
```

#### 5. STANDARD DEVIATION

```python
standard_deviation = np.std(energy_consumption)
print(standard_deviation
     )
```

#### 6. RESHAPING

```python
energy_consumption2 = np.array([1200 , 3400 , 2900, 1800, 2500 , 1800])

# for reshaping the size of array should be same which is = 6
reshaped = energy_consumption2.reshape((3,2))
reshaped2 = energy_consumption2.reshape((6,1))
reshaped3 =  energy_consumption2.reshape((2,3))

print("The new reshaped" , reshaped)
print("The new reshaped" , reshaped2)
print("The new reshaped" , reshaped3)

# flatteining the array --> means converting nD array into 1D 
print("The flatten array is:",reshaped.flatten())

```

#### 7. TRANSPOSE

```python
# Tranpose 
print("Tranpose of array is: ", reshaped3.T)
```

#### 8. RESIZING

```python
# resizing the array

energy_consumption2 = np.array([1200 , 3400 , 2900, 1800, 2500 , 1800])
resized = np.resize(energy_consumption2 , (1,9))
print(resized)
```

#### 9. OTHER NUMPY METHODS

```python
# genrating random integer
print(np.random.randint(0 , 10, 1))

# multiplication of two array
print("The self multiplication of energy consumption array is :",np.dot(energy_consumption2 , energy_consumption2))

# power ele by 2
l = [2, 4, 5, 6]
arr = np.array(l)
print("The square of each element from the array: " , arr, "is" , np.power(arr, 2))

# square root of an elements of array
print("The square root of each element from the array: " , arr, "is" ,np.sqrt(arr))

# max element from array
print("The max element from the array: " , arr, "is" , np.max(arr))
      
# Creates an array with evenly spaced values from 0 to 9
print(np.arange(10))
```
