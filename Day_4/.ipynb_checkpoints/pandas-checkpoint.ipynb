```python
import pandas as pd

renewable_sources = ["Solar", "Wind", "Hydropower",  "Geothermal", "Biomass"]

# creating series
renewable_series = pd.Series(renewable_sources)
# print("Renewable Energy Sources:")
print(renewable_series)

# creatind data frame
data = {
    "Project": ["Solar Farm A", "Wind Turbine X", "Hydropower Y", "Solar Roof Z", "Geothermal Plant P"],
    "Technology": ["Solar", "Wind", "Hydropower", "Solar", "Geothermal"],
    "Capacity (MW)": [150, 300, 200, 50, 100],  # Megawatts
    "Cost (Million $)": [200, 400, 350, 100, 250],  # Project cost
    "Location": ["California", "Texas", "Washington", "Nevada", "Idaho"],
    "Completion Year": [2023, 2024, 2022, 2025, 2023]
}

frame = pd.DataFrame(data)
print(frame)


# top 3
print("The Top 3 data are:")
print(frame.head(3))

# last 3
print("The last 3 data are:")
print(frame.tail(3))

# slicing
frame[3:5]

# giving the datatypes of columns
print("The data types are : " , frame.dtypes)

# return the shape of tables/dataframe
print("The size of data frame is :" , frame.shape)

```

```python
# Adding new features (column) / Feature enginnering
frame["Cost per MW"] = frame["Cost (Million $)"] / frame["Capacity (MW)"]
frame.head()
```

```python
# grouping
grouped_data = frame.groupby("Technology")["Capacity (MW)"].sum()
print(grouped_data)
```
