### ADVANCE VISUALIZATION

```python
# heat map

correlation_matrix = data.drop(columns="species").corr()

# plot heatmap
slt.heatmTap(correlation_matrix , annot = True , cmap = "coolwarm" , linewidth = 0.5)

# add title
plt.title("Correlation")
plt.show()
```

```python
# box plot
slt.boxplot(data)

plt.title("Distribution of sepal length")
plt.show()
```
