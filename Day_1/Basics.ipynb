```python
Weekly_Temp = [25 ,45 , 89, 36, 12, 42, 18 ]

#removing the element
Weekly_Temp.remove(42)

#adding at end
Weekly_Temp.append(33)

#inserting at any location
Weekly_Temp.insert(1, 17)

print(f"Weekly Temperature: {Weekly_Temp}")
```

DICTIONARY OPERATIONS

```python
user = {
    "name" : "Rohit",
    "email" : "abc@gmail.com",
    "phone" : 9767272177,
    "age" : 21,
    "salary" : 2356.12
}

#adding elements
user["city"] = "Pune"

#deleting elements
del user["age"]

print(f"User Deatils: {user}")

#returning keys from dictionary
user.keys()

#returning values from dictionary
user.values()

# NOTE:- from keys() and values() it will return only  response with 1 values because it override with the two functions
```
