## CLASS

```python
class EnergySystem :

    # __init__ method is  used to setup the object whenever it is created
    def __init__(rohit , building_name , energy_consumption, emmision_factor):

        # Initalizing the attributes(variables)
        rohit.building_name = building_name
        rohit.energy_consumption = energy_consumption
        rohit.emmision_factor = emmision_factor

        # method to calculate the carbon footprint
    def calculate_carbon_footprint(rohit):
        return rohit.energy_consumption * rohit.emmision_factor

        # method to calculate the energy savings
    def calculate_energy_savings(self):
        return self.energy_consumption * 0.10 

# creating a instance of a class
building1 = EnergySystem("Building A" , 5000 , 0.45)

# calling the methods from the class
carbon_footprint = building1.calculate_carbon_footprint()
energy_savings = building1.calculate_energy_savings()

# accesing the attributes from the class
print(f"{building1.building_name} : {building1.energy_consumption} kWh , {building1.emmision_factor}")
print(f"Carbon footprint = {carbon_footprint} kg CO2")
print(f"Energy Savings = {energy_savings} kWh")


#Inheritance
class SolarEnergySystem(EnergySystem):
    def __init__(self , building_name , energy_consumption, emmision_factor , solar_production ):

        # Inheriting the property of parent class 
        # this method is used to setup the attributes when the object is created 
        super().__init__(building_name, energy_consumption, emmision_factor)
        self.solar_production = solar_production

    # calculating the net energy conssumption
    def net_energy_consumption(self):
        return self.energy_consumption - self.solar_production

solar_building = SolarEnergySystem("Solar Building" , 5000 , 0.45 , 1500)

net_consumption = solar_building.net_energy_consumption()

print(f"Net energy consumption {net_consumption} kWh")
```

## ENCAPSULATION

```python
class EnergySystem :

    # __init__ method is  used to setup the object whenever it is created
    def __init__(rohit , building_name , energy_consumption, emmision_factor):

        # Initalizing the attributes(variables)
        rohit.building_name = building_name
        rohit._energy_consumption = energy_consumption #protected variable
        rohit._emmision_factor = emmision_factor ##protected variable

    # method to acces energy consumption
    def get_protected_attribute(self):
        return self._energy_consumption

        # method to calculate the carbon footprint
    def calculate_carbon_footprint(rohit):
        return rohit._energy_consumption * rohit._emmision_factor

        # method to calculate the energy savings
    def calculate_energy_savings(self):
        return self._energy_consumption * 0.10 

# creating a instance of a class
building1 = EnergySystem("Building A" , 5000 , 0.45)

# calling the methods from the class
carbon_footprint = building1.calculate_carbon_footprint()
energy_savings = building1.calculate_energy_savings()

# accesing the attributes from the class
print(f"{building1.building_name} : {building1._energy_consumption} kWh , {building1._emmision_factor}")
print(f"Carbon footprint = {carbon_footprint} kg CO2")
print(f"Energy Savings = {energy_savings} kWh")

#printing the the protected attribute
print(f"Energy Consumption {building1.get_protected_attribute()} kWh")
```

## POLYMORPHISM

```python
# POLYMORPHISM

# parent class
class EnergySystem :

    # __init__ method is  used to setup the object whenever it is created
    def __init__(rohit , building_name , energy_consumption, emmision_factor):

        # Initalizing the attributes(variables)
        rohit.building_name = building_name
        rohit.energy_consumption = energy_consumption
        rohit.emmision_factor = emmision_factor

        # method to calculate the carbon footprint
    def calculate_carbon_footprint(rohit):
        return rohit.energy_consumption * rohit.emmision_factor

        # method to calculate the energy savings
    def calculate_energy_savings(self):
        return self.energy_consumption * 0.10 

# child class
class SolarEnergySystem(EnergySystem):
    def __init__(self , building_name , energy_consumption, emmision_factor , solar_production ):

        # Inheriting the property of parent class 
        # this method is used to setup the attributes when the object is created 
        super().__init__(building_name , energy_consumption, emmision_factor)
        self.solar_production = solar_production

    # calculating the net energy conssumption
    def net_energy_consumption(self):
        return self.energy_consumption - self.solar_production

    def calculate_carbon_footprint(self):
        net_consumption = self.net_energy_consumption()
        return net_consumption * self.emmision_factor
    

solar_building = SolarEnergySystem("Solar Building" , 5000 , 0.45 , 1500)

# net_consumption = net_energy_consumption()
carbon_footprint = solar_building.calculate_carbon_footprint()

print(f"Carbon Footprint : {carbon_footprint} kWh")
```
