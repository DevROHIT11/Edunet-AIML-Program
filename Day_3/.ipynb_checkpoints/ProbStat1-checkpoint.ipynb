```python
# parent class
class EnergyResources:

    def __init__(self , Energy_name , factor ):
        self.Energy_name = Energy_name
        self.factor = factor

    def calculate_energy(self):
        _total_power = self.capacity * self.factor
        return _total_power

# solar energy
class SolarEnergy(EnergyResources) :
    
    def __init__(self , capacity):
        super().__init__("Solar" , 0.4 )
        self.capacity = capacity

# wind energy
class WindEnergy(EnergyResources) :
    
    def __init__(self , capacity):
        super().__init__("Wind" , 0.2)
        self.capacity = capacity

# hydro energy
class HydroEnergy(EnergyResources) :
    def __init__(self , capacity):
        super().__init__("Hydro" , 0.3)
        self.capacity = capacity

# instances
solar = SolarEnergy(200)
wind = WindEnergy(500)
hydro = HydroEnergy(300)

# calculating the enrgy
solar_power = solar.calculate_energy()
wind_power = wind.calculate_energy()
hydro_power = hydro.calculate_energy()

#display values
print(f"Solar energy = {solar_power}")
print(f"Wind energy = {wind_power}")
print(f"Hydro energy = {hydro_power}")

# decision engine
if(solar_power > wind_power and solar_power > hydro_power ):
    print("Use Solar energy")
elif(wind_power > hydro_power ):
    print("Use Wind energy")
else:
    print("Use Hydro energy")
```
